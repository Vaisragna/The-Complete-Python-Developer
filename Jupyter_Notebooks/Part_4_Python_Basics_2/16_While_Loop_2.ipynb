# While Loop 2

```python
my_list = [1,2,3]
```

```python
for item in my_list:
    print(item)
```

```python
i = 0
while i < len(my_list):
    print(my_list[i])
    i += 1
```

```python
while True:
    input('say something: ')
    break
```

```python
while True:
    response = input('say something: ')
    if (response == 'bye'):
        break
```
