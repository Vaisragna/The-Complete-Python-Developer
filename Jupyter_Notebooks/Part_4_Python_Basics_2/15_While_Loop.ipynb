# While Loop

A while loop in Python is a control structure that repeats a block of code while a condition is true (True).

In other words, the program continues executing the code until the condition is no longer satisfied.

```python
i = 0
while i < 5:
    print (i)
    i = i + 1
```

```python
i = 0
while i < 5:
    print (i)
    i += 1
```

```python
i = 0
while i < 5:
    print (i)
    i += 1
else:
    print('done with all the work')
```
