# break, continue, pass

1. break immediately stops a loop (either for or while) and the program continues with the code after the loop.


```python
my_list = [1,2,3]
for item in my_list:
    print(item)
    break
```

2. continue skips the rest of the current iteration and moves to the next iteration of the loop.

```python
my_list = [1,2,3]
for item in my_list:
    continue
    print(item)
```

3. pass does nothing. It is used as a placeholder when a statement is required syntactically but you don't want to execute any code yet.

```python
my_list = [1,2,3]
for item in my_list:
    # im thinking on something
    pass
```
