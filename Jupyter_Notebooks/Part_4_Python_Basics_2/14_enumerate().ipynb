# Enumerate()

enumerate() is a built-in Python function used when you want to loop over something and also keep track of the index (position) of each item.

Instead of manually creating a counter variable, enumerate() gives you the index and the value at the same time.

```python
# Char
for i,char in enumerate('Heloooo'):
    print (i,char)
```

```python
# List
for i,char in enumerate([1,2,3]):
    print (i,char)
```

```python
# Tuple
for i,char in enumerate((1,2,3)):
    print (i,char)
```

```python
for i,char in enumerate(list(range(100))):
    if char == 50:
        print(f'index of 50 is: {i}')
```
